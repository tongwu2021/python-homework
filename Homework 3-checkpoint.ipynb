* length(X)를 다음과 같이 정의한다 : 10진수로 표현된 X의 자릿수
* Strange라는 숫자를 아래과 같이 정의한다.
    * 음이 아닌 정수 중 자릿수가 1인 모든 숫자는 Strange이다.
    * length(X) > 1 인 숫자 X 는 다음 2가지를 만족하면 Strange이다.
        * X는 length(X)로 정확히 나누어 떨어진다.
        * X / length(X) 가 strnage이다.

* 이 문제에서 숫자 하나가 주어질 때, 이 숫자가 Strange인지 판단하는 함수 isStrange(num)을 작성하시오.
* isStrange(num) 함수가 재귀적인 함수가 되도록 작성해 보세요.

* [예시]
    1. num = 0                      =>    True
    2. num = 56                     =>    True
    3. num = 100                    =>    False
    4. num = 7882602506492313600    =>    True
    5. num = 6012302549522098651    =>    False

````python
def isStrange(num):
    if num >= 0 and num < 10:
        return True
    elif len(str(num)) > 1:
        if (num % len(str(num)) == 0 and isStrange(int(num/len(str(num))))):
            return True
        else:
            return False
    else:
        return False

print(isStrange(0))
print(isStrange(56))
print(isStrange(100))
print(isStrange(7882602506492313600))
print(isStrange(6012302549522098651))
````

num이 0보다 크거나 같고 10보다 작으면 True를 반환하고, 만약 자릿수가 1을 넘길 때, num이 자릿수에 딱 맞게 떨어지고 재귀를 사용한 isStrange() 함수에 'num/len(str(num)'을 넣어 True가 나올 경우 최종적으로 True를 반환한다. 나머지는 전부 False를 반환한다. 여기서 'int(num/len(str(num))'를 사용한 것은 int형으로 입력받게 하기 위함이다.


````python
def solve(a):
    sum_ = 0
    max_n = 0
    temp = 0
    result = []

    for i in range(len(a)):
        sum_ += a[i]
        if(max_n < a[i]): max_n = a[i]

    for i in range(len(a)):
        temp += a[i]
        if max_n <= temp and sum_ % temp == 0 :
            result.append(temp)

    return result
a = [1, 2, 1, 1, 1, 2, 1, 3]
print(solve(a))
a = [1, 1, 1, 1]
print(solve(a))
a = [9999] * 10**5
print(solve(a))
````

a의 목록에서 for문을 이용해 총합을 만들기 위해 sum_에 일일이 더하는데 sum() 함수를 쓰지 않는 이유는 a의 인덱스 중에서 가장 큰 값을 찾기 위함이다.<br>
그리고 두 번째 for문에서는 temp 또한 sum_과 마찬가지로 하고 만약 temp가 max_n보다 크고 sum_을 temp로 나눴을 때 나머지가 없으면 result 배열에 추가한다.<br>
'sum_ % temp == 0'는 버스에 인원이 딱 떨어지는 지를 확인하고 'max_n <= temp'는 버스 크기가 인덱스에서 가장 큰 값보다 크거나 같아야 되고 temp를 일일이 더하는 이유는 앞의 그룹들이 차례대로 출발해야 하기에 순서를 충족할 수 있는 조건을 맞추기 위해서이다.

* 요구사항 2. 순수 텍스트 안에 존재하는 단어들에 대해 string 모듈을 활용하여 모든 punctuation (구두문자)를 완벽히 제거하시오.
    * 예: ['world!', ':', '+hello+', '~python$$$', '=', 'world'] ---> ['world', 'hello', 'python', 'world']
    * 모든 punctuation을 확인하는 방법은 아래와 같다.
```python
import string
print(string.punctuation)
```
* 요구사항 3. 만약 punctuation (구두문자)를 제거한 이후 공백으로 판명되는 단어들은 해당 단어 리스트에서 제거하시오.
    * 예: ['컴퓨터공학부!', '---?', '%지향~', '...'] ---> ['컴퓨터공학부', '', '지향', ''] ---> ['컴퓨터공학부', '지향']

* 요구사항 3. 각 단어들의 출현빈도를 사전형태로 저장하여 출력하시오.
    * 예: ['world', 'hello', 'python', 'world', '컴퓨터공학부', '지향'] ---> {'world': 2, 'hello': 1, 'python': 1, '컴퓨터공학부': 1, '지향': 1}

* 요구사항 4. 다음 5개의 사이트에 대한 단어 출현빈도 결과를 출력하시오.
    * http://cse.koreatech.ac.kr
    * https://www.koreatech.ac.kr
    * http://www.naver.com
    * http://www.daum.net
    * http://www.nytimes.com

````python
import requests
import string

req = requests.get('http://www.koreatech.ac.kr')
source = req.text

source = source.strip()
word = ''
check = True
result = []
result2 = []
index = 0

for i in range(len(source)):
    if (source[i] == '>' and check == False):
            check = True
            continue
    if (source[i] == '<'):
            check = False
            continue
    if(check == True):
        word += source[i]

result = word.split()

for i in range(len(result)): 
    result[i] = ''.join(a for a in result[i] if a not in string.punctuation)
result = ' '.join(result).split()
result.sort()
dict_result = {}

for i in result:
    if i in dict_result:
        dict_result[i] += 1
    else:
        dict_result[i] = 1
print(dict_result)
````

받아온 HTML에서 >가 나오고 check이 false이면 check을 True로 만들고 <가 나오면 check을 False로 만든다. 그래서 check이 True인 경우 그 문자열을 word에 넣는다. 이렇게 되면 > < 사이의 문자열이 word에 들어가게 된다.<br>
그리고 공백을 기준으로 첫 번째 결과를 리스트에 나눠 넣는다면 순수한 단어를 찾는 첫 번째 단계가 완성된다.<br>
이 다음은 구두문자를 제거하기 위해 string 모듈을 활용하여 for문과 join을 이용해 각 인덱스의 목록에 구두 문자가 있는 내용이 있으면 문자열 변경 과정에서 추가하지 않는다.<br>
그 후 공백을 기준으로 다시 단어들을 분리하여 리스트 형태로 만들고 sort()를 이용해 정렬한다.<br>
